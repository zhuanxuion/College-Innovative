```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取CSV文件
filename = 'log1.csv'  # 确保这里的文件名与你的CSV文件名一致
df = pd.read_csv(filename)
```

```python
# 提取epoch、Train_loss和Test_loss列
epochs = df['epoch']
train_losses = df['Train_loss']
test_losses = df['Test_loss']

# 创建图表
plt.figure(figsize=(10, 6))

# 绘制Train_loss曲线
plt.plot(epochs, train_losses, label='Train Loss', marker='o')

# 绘制Test_loss曲线
plt.plot(epochs, test_losses, label='Test Loss', marker='x', linestyle='--')

# 添加标题和坐标轴标签
plt.title('Training and Testing Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# 添加图例
plt.legend()

# 显示网格
plt.grid(True)
plt.savefig("loss.png")
# 显示图表
plt.show()
```

```python

# 提取epoch、Train_auroc和Test__auroc列
epochs = df['epoch']
Train_auroc = df['Train_auroc']
test_losses = df['Test_auroc']

# 创建图表
plt.figure(figsize=(10, 6))


plt.plot(epochs, Train_auroc, label='auroc', marker='o')


plt.plot(epochs, test_losses, label='Test_auroc', marker='x', linestyle='--')

# 添加标题和坐标轴标签
plt.title('Training and Testing Train_auroc over Epochs')
plt.xlabel('Epoch')
plt.ylabel('auroc')

plt.ylim(0.8,1)


# 添加图例
plt.legend()

# 显示网格
plt.grid(True)
plt.savefig("auroc.png")
# 显示图表
plt.show()
```

```python
# 提取epoch、Train_auroc和Test__auroc列
epochs = df['epoch']
Train_f1_score = df['Train_f1_score']
test_f1_score = df['Test_f1_score']

# 创建图表
plt.figure(figsize=(10, 6))


plt.plot(epochs, Train_f1_score, label='Train_f1_score', marker='o')


plt.plot(epochs, test_f1_score, label='Test_f1_score', marker='x', linestyle='--')

# 添加标题和坐标轴标签
plt.title('Training and Testing f1_score over Epochs')
plt.xlabel('Epoch')
plt.ylabel('f1_score')


# 添加图例
plt.legend()

# 显示网格
plt.grid(True)
plt.savefig("f1.png")

# 显示图表
plt.show()
```
